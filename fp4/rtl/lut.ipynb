```python



e2m1_input_range = {.0, .5, 1.0, 1.5, 2.0, 3.0, 4.0, 6.0}
e2m1_input_map = {
    .0: "000", 
    .5: "001",
    1.0: "010",
    1.5: "011",
    2.0: "100",
    3.0: "101",
    4.0: "110",
    6.0: "111",
}

output_map = {}

# Generate output map for multiplication values
for a in e2m1_input_map.keys():
    for b in e2m1_input_map.keys():
        lutInputEnc = e2m1_input_map[a] + e2m1_input_map[b]
        intValue = int(a / 0.5) * int(b/0.5)
        output_map[lutInputEnc] = intValue
        # print(f"{a} * {b} => 6'b{lutInputEnc} : 8'b{intValue:08b} ({intValue/4})")
        print(f"8'b{intValue:08b}, /* {a} * {b} => {intValue/4} */ ")
```

```python
print(output_map)
print(len(output_map)/256)
```

```python

minterms = [set() for _ in range(0, 8)]
maxterms = [set() for _ in range(0, 8)]

for input_pattern, output_pattern in output_map.items():
    for i in range(0, 8):
        if (output_pattern & (1 << i)) != 0:
            minterms[i].add(input_pattern)
        else:
            maxterms[i].add(input_pattern)
            

print("Minterms:")
for i, input_patterns in enumerate(minterms):
    print(f"out[{i}]: {input_patterns}")

print("=" * 20)

print("Maxterms:")
for i, input_patterns in enumerate(maxterms):
    print(f"~out[{i}]: {input_patterns}")
```
